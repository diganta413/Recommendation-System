```python
import pandas as pd
import numpy as np
```

```python
d=pd.read_csv("movies.csv")
d.head()
```

```python
df=d[["imdb_title_id","original_title","genre"]]
df.head()
```

```python
df["genre"]=df.genre.str.split(",")
df.head()
```

```python
movies=df.copy()
for index,row in df.iterrows():
    for i in row["genre"]:
        movies.at[index,i]=1
movies=movies.fillna(0)
movies.head()
#movies.shape
```

```python
user_input={"original_title":["Miss Jerry","The Story of the Kelly Gang","Den sorte drøm","Cleopatra","L'Inferno"],"Rating":[5.5,2.2,3.3,4.5,7.8]}
user=pd.DataFrame(user_input)
user
```

```python
input_id=df[df["original_title"].isin(user["original_title"].tolist())]
input_id=input_id.drop([input_id.index[5],input_id.index[6],input_id.index[7],input_id.index[8]])
input_id
movie_input=pd.merge(input_id,user)
movie_input=movie_input.drop("genre",axis="columns")
movie_input
```

```python
final_user_input=movies[movies["imdb_title_id"].isin(movie_input["imdb_title_id"].tolist())]
final_user_input=final_user_input.drop(["genre","imdb_title_id","original_title"],axis="columns")
final_user_input
```

```python
rating_table=movie_input[["Rating"]]
rating_table
```

```python
user_profile=final_user_input.transpose().dot(movie_input["Rating"])
user_profile
```

```python
movies_table=movies.copy()
movies_table=movies.set_index(movies["imdb_title_id"])
movies_table=movies_table.drop(["imdb_title_id","original_title","genre"],1)
movies_table.head()
```

```python
recommendation_table=((movies_table*user_profile).sum(axis=1))/(user_profile.sum())
recommendation_table.head()
```

```python
recommendation_table=recommendation_table.sort_values(ascending=False)
recommendation_table.head()
```

```python
recomended_movie_list=df[df["imdb_title_id"].isin(recommendation_table.head(20).keys()).tolist()]
recomended_movie_list
```
